# Regular expressions

Regular expressions provide a flexible way to search or match string patterns in text
<br>Python’s built-in re module is responsible for applying regular expressions to strings.
<br>The re module functions fall into three categories: pattern matching, substitution, and splitting

```python
print("c:\new\test")
```

```python
print(r"c:\new\test")  # raw string
```

```python
import re
```

```python
text="Meier Maier Maayer MYER mYER meyer Maaaaaaaaaayer Myer Maeyer"
```

Square brackets, "[" and "]", are used to include a character class.
<br>[xyz] means e.g. either an "x", an "y" or a "z".
<br>[a-e] a simplified writing for [abcde] or [0-5] denotes [012345].

```python
pattern="M[ae]{1,2}[iy]er"  # raw-string
```

```python
result=re.findall(pattern,text)
```

```python
result
```

## Matching any character 
• The ”.” character is used to match any character.

```python
text="abc akc adc adbc"
```

```python
pattern=r"a.c"
```

```python
re.findall(pattern,text)
```

## Matching multiple copies

```python
text="ABC12 DEK222 FGHLY 1010"
```

from above text find only alphabetic characters

```python
pattern=r"[A-Za-z]"
```

```python
re.findall(pattern,text)
```

```python
pattern=r"[A-Za-z]{5}"
```

```python
re.findall(pattern,text)
```

if you want to get only integers use \d

```python
text=text+" TET168"
```

```python
text
```

```python
pattern=r"\d"
```

```python
re.findall(pattern,text)
```

```python
pattern=r"[A-Z]{3}\d{3}"
```

```python
re.findall(pattern,text)
```

# Other functions

Re.sub() finds all substrings where the RE matches, and replace them with a different string

```python
re.sub(r"a","*","ababababa")
```

Re.split() Split the string into a list, splitting it wherever the RE matches

```python
re.split(r"a","ababababa")
```
